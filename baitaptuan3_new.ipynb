```python
def dao_nguoc_bang_chuoi(n):
    """
    Cách 1: Sử dụng xử lý chuỗi (String Slicing) - Rất nhanh trong Python
    """
    # Xử lý dấu âm nếu có
    if n < 0:
        # Lấy giá trị tuyệt đối, chuyển thành chuỗi, đảo ngược [::-1], rồi ép lại kiểu int và thêm dấu trừ
        return -int(str(abs(n))[::-1])
    else:
        return int(str(n)[::-1])

def dao_nguoc_bang_toan_hoc(n):
    """
    Cách 2: Sử dụng toán học (Modulo và Chia nguyên) - Tư duy thuật toán chuẩn
    """
    so_am = n < 0
    n = abs(n) # Tạm thời làm việc với số dương
    so_dao_nguoc = 0
    
    while n > 0:
        chu_so_cuoi = n % 10                # Phép chia lấy dư: Bóc chữ số ngoài cùng bên phải
        so_dao_nguoc = so_dao_nguoc * 10 + chu_so_cuoi # Dịch số cũ sang trái 1 hàng (nhân 10) rồi cộng số mới
        n = n // 10                         # Phép chia nguyên: Cắt bỏ chữ số vừa lấy
        
    # Viết rõ ràng khối lệnh kiểm tra dấu thay vì dùng cú pháp viết tắt
    if so_am == True:
        # Nếu mẫu vật ban đầu mang dấu âm, ta phải "mặc lại áo" (thêm dấu trừ) cho nó
        return -so_dao_nguoc
    else:
        # Nếu mẫu vật ban đầu là số dương, cứ trả về nguyên bản
        return so_dao_nguoc

# ==========================================
# CHƯƠNG TRÌNH CHÍNH
# ==========================================
try:
    # Nhập dữ liệu từ người dùng
    so_n = int(input("Nhập vào một số nguyên n: "))
    
    print("\n--- KẾT QUẢ ---")
    print(f"Số ban đầu: {so_n}")
    
    # Bạn có thể dùng 1 trong 2 cách, ở đây mình in ra cả 2 để bạn so sánh
    ket_qua_c1 = dao_nguoc_bang_chuoi(so_n)
    print(f"Kết quả (Cách dùng chuỗi): {ket_qua_c1}")
    
    ket_qua_c2 = dao_nguoc_bang_toan_hoc(so_n)
    print(f"Kết quả (Cách dùng toán học): {ket_qua_c2}")

except ValueError:
    print("Lỗi: Vui lòng chỉ nhập số nguyên hợp lệ!")




#tim so hang don vi, hang nghin cua mot so nguyen duong
n = int(input("Nhập vào một số nguyên dương n: "))
if n < 0:
    print("Lỗi: Vui lòng nhập số nguyên dương!")
else:
    hang_don_vi = n % 10
    hang_nghin = (n // 1000) % 10
    print(f"Số hàng đơn vị của {n} là: {hang_don_vi}")
    print(f"Số hàng nghìn của {n} là: {hang_nghin}")

```
