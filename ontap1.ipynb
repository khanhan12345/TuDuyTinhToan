no1 taxi fee

```python
def smart_taxi_fare(car_type,start_hour,km,wait_minutes):
    import math
    total=0
    if 0<=km<=0.8:
        if car_type==4:
            total=total+11000
        if car_type==7:
            total=total+12000
    if 0.8<km<=20:
        if car_type==4:
            total=11000+(km-0.8)*13000
        if car_type==7:
            total=12000+(km-0.8)*14000
    if km>20:
        if car_type==4:
            total=11000+(20-0.8)*13000+(km-20)*10500
        if car_type==7:
            total=12000+(20-0.8)*14000+(km-20)*12000
    if wait_minutes>5:
        total=total+math.ceil((wait_minutes-5)/5)*2000
    if 22<=start_hour<=23 or 0<=start_hour<=5:
        total=total*1.2
    if km>30 and 6<=start_hour<22:
        total=total*0.9
    if total<50000 and km>5:
        return 50000
    return int(total)
car_type=int(input('4 or 7'))
start_hour=float(input('enter start hour(s): '))
km=float(input('enter km: '))
wait_minutes=float(input('enter wait minute(s): '))
print(smart_taxi_fare(car_type,start_hour, km, wait_minutes))
        
    
```

no2 given x, return true if there is n>=1: x=1+2+3+...+n, otherwise return false

```python
x=int(input())

def check(x):
    count=0
    for n in range(1,x):
        count=count+n
        if x==count:
            return True
    return False
print(check(x))
    
    

```

no3 cho toa do (x,y) cua ba diem a,b,c, kiem tra tam giac (kiem tra 3 diem co thang hang hay khong) / (kiem tra vecto chi phuong)

```python
x1,y1=map(float, input('nhap x1, y1 cua diem a: ').split())
x2,y2=map(float, input('nhap x2, y2 cua diem b: ').split())
x3,y3=map(float, input('nhap x3, y3 cua diem c: ').split())
ab_x=x2-x1
ab_y=y2-y1
ac_x=x3-x1
ac_y=y3-y1
def triangle_check(ab_x,ab_y,ac_x,ac_y):
    if ab_x/ac_x != ab_y/ac_y:
        return False
    return True
print(triangle_check(ab_x,ab_y,ac_x,ac_y))
```
