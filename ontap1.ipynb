no1 taxi fee

```python
def smart_taxi_fare(car_type,start_hour,km,wait_minutes):
    import math
    total=0
    if 0<=km<=0.8:
        if car_type==4:
            total=total+11000
        if car_type==7:
            total=total+12000
    if 0.8<km<=20:
        if car_type==4:
            total=11000+(km-0.8)*13000
        if car_type==7:
            total=12000+(km-0.8)*14000
    if km>20:
        if car_type==4:
            total=11000+(20-0.8)*13000+(km-20)*10500
        if car_type==7:
            total=12000+(20-0.8)*14000+(km-20)*12000
    if wait_minutes>5:
        total=total+math.ceil((wait_minutes-5)/5)*2000
    if 22<=start_hour<=23 or 0<=start_hour<=5:
        total=total*1.2
    if km>30 and 6<=start_hour<22:
        total=total*0.9
    if total<50000 and km>5:
        return 50000
    return int(total)
car_type=int(input('4 or 7'))
start_hour=float(input('enter start hour(s): '))
km=float(input('enter km: '))
wait_minutes=float(input('enter wait minute(s): '))
print(smart_taxi_fare(car_type,start_hour, km, wait_minutes))
        
    
```

```python
x=int(input())

def check(x):
    count=0
    for n in range(1,x):
        count=count+n
        if x==count:
            return True
    return False
print(check(x))
    
    

```

```python
a=float(input())
b=float(input())
c=float(input())
def caculus(a,b,c):
    nghiem=[]
    if a==0 and b!=0:
        nghiem.append(-c/b)
        return nghiem
    if a==0 and b==0:
        return nghiem
    if a!=0:
        d=b*b-4*a*c
        if d<0:
            return nghiem
        if d==0:
            nghiem.append(-b/(2*a))
            return nghiem
        if d>0:
            r1=(-b+d**0.5)/(2*a)
            r2=(-b-d**0.5)/(2*a)
            nghiem.append(r1)
            nghiem.append(r2)
            return nghiem
print(caculus(a,b,c))
```

```python
d1,m1,y1=map(int, input('nhap ngay thang nam sinh nguoi 1: ').split())
d2,m2,y2=map(int, input('nhap ngay thang nam sinh nguoi 2: ').split())
def birthday(d1,m1,y1,d2,m2,y2):
    if y1>y2:
        return 1
    if y1<y2:
        return 2
    if y1==y2:
        if m1>m2:
            return 1
        if m1<m2:
            return 2
        if m1==m2:
            if d1>d2:
                return 1
            if d1<d2:
                return 2
            if d1==d2:
                return 0
print(birthday(d1,m1,y1,d2,m2,y2))
```

no16 viet ham tra ve goc abc theo don vi do khi biet toa do diem a, diem b va diem c

```python
x1,y1=map(float, input('nhap toa do diem a').split())
x2,y2=map(float, input('nhap toa do diem b').split())
x3,y3=map(float, input('nhap toa do diem c').split())
import math
bax=x1-x2
bay=y1-y2
bcx=x3-x2
bcy=y3-y2
tichvohuong=bax*bcx+bay*bcy
ba=(bax**2+bay**2)**0.5
bc=(bcx**2+bcy**2)**0.5
tichdodai=ba*bc
cosbeta=tichvohuong/tichdodai
def beta(cosbeta):
    goc_radian=math.acos(cosbeta)
    return math.degrees(goc_radian)

print(beta(cosbeta))
```

no29 viet ham is on prodcut of other (a,b,c) tra ve true neu mot trong ba so bang tich hai so con lai

```python
#se co the bao loi khi du lieu dau vao la float->stupid computer
def is_one_product_of_others(a,b,c):
    lamtrona=math.isclose(a,b*c)
    lamtronb=math.isclose(b,a*c)
    lamtronc=math.isclose(c,a*b)

    return lamtrona or lamtronb or lamtronc
#assert is_one_product_of_others(2,3,6)==False, 'ok'
assert is_one_product_of_others(0.1,0.2,0.02)==False, 'ok'
```

```python
def detect_number_type(s):
    import math
    if not s:
        return 'neither'
    check_s=s
    if s[0]=='+' or s[0]=='-':
        check_s=s[1:]
    if not check_s:
        return 'neither'
    if check_s.isdigit():
        return 'int'
    if '.' in check_s:
        parts=check_s.split('.')
        print(parts)
        if len(parts)==2 and parts[0].isdigit() and parts[1].isdigit():
            return 'float'
    return 'neither'

assert detect_number_type('123.345.21')=='neither', 'sai roi'
    
```
