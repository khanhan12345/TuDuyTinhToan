bai 3 mod and div
//: chia lay nguyen
%: chia lay du
**: luy thua

```python
#bai 3
a=int(input())
b=int(input())
#a, b=map(int, input().split())
tong = a+b
hieu=a-b
tich=a*b
thuong=a/b
phan_nguyen=a//b
phan_du=a%b
print(f"{a} +{b} = {tong}")
print(f"{a} -{b} = {hieu}")
print(f"{a} *{b} = {tich}")
print(f"{a} /{b} = {thuong}")
print(f"{a} //{b} = {phan_nguyen}")
print(f"{a} %{b} = {phan_du}")






#print(a+b, a-b, a/b, a*b, a//b, a%b, sep= "\n")
```
