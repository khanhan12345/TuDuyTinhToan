bai 8 giai he pt 2 an

```python
def giai_he_phuong_trinh(a, b, m, c, d, n):
    """
    Hàm giải hệ phương trình tuyến tính 2 ẩn:
    ax + by = m
    cx + dy = n
    """
    
    # Bước 1: Tính các định thức (Determinants)
    D = a * d - b * c
    Dx = m * d - b * n
    Dy = a * n - m * c

    # Bước 2: Kiểm tra các điều kiện (Logic nhánh)
    if D != 0:
        # Trường hợp 1: Có nghiệm duy nhất
        x = Dx / D
        y = Dy / D
        return f"Hệ phương trình có nghiệm duy nhất:\nx = {x}\ny = {y}"
        
    else:
        # Trường hợp D = 0
        if Dx == 0 and Dy == 0:
            # Trường hợp 2: Vô số nghiệm
            return "Hệ phương trình có VÔ SỐ nghiệm (hai đường thẳng trùng nhau)."
        else:
            # Trường hợp 3: Vô nghiệm
            return "Hệ phương trình VÔ NGHIỆM (hai đường thẳng song song)."
a, b, m, c, d, n = map(float, input().split())
print(giai_he_phuong_trinh(a, b, m, c, d, n))
```
