w4a6 prime number check

```python
def check_prime(n):
    if n<2:
        return False
    gioihan=int(n**0.5)
    for i in range(2, gioihan+1):
        if n%i==0:
            return False
    return True


a=int(input('nhap a: '))
b=int(input('nhap b: '))
for so in range(a, b+1):
    if check_prime(so):
        print(so, end=" ")
```

w4a7 uoc so nguyen to lon nhat i love gemini

```python
def tim_uoc_nguyen_to_lon_nhat(n):
    uoc_lon_nhat = -1
    
    # Bước 1: "Vắt kiệt" số 2 (số nguyên tố chẵn duy nhất)
    # Tách riêng số 2 để lát sau ta chỉ cần kiểm tra các số lẻ (tăng tốc độ gấp đôi)
    while n % 2 == 0:
        uoc_lon_nhat = 2
        n = n // 2  # Dùng // để chia lấy nguyên
        
    # Bước 2: Kiểm tra các số lẻ từ 3 trở đi.
    # Tối ưu thuật toán: Chỉ cần chạy đến căn bậc 2 của n là đủ.
    # Vì nếu n có 1 ước số > căn(n), thì chắc chắn nó phải nhân với 1 số < căn(n) (mà số này ta đã xử lý rồi)
    i = 3
    while i * i <= n:
        while n % i == 0:
            uoc_lon_nhat = i
            n = n // i
        # Bỏ qua các số chẵn, bước nhảy là 2 (3, 5, 7, 9...)
        i += 2
        
    # Bước 3: Trường hợp đặc biệt
    # Nếu sau khi chia hết vòng lặp mà phần còn lại của n vẫn lớn hơn 2
    # Thì bản thân phần còn lại đó chính là một số nguyên tố (và là số lớn nhất)
    if n > 2:
        uoc_lon_nhat = n
        
    return uoc_lon_nhat

# --- PHẦN CHƯƠNG TRÌNH CHÍNH ---
try:
    # Nhận đầu vào
    n = int(input())
    
    # Gọi hàm và in kết quả
    if n >= 2:
        ket_qua = tim_uoc_nguyen_to_lon_nhat(n)
        print(ket_qua)
    else:
        print("Vui lòng nhập n >= 2 theo đúng đề bài!")
except ValueError:
    pass # Bỏ qua lỗi nếu nhập không phải là số (tùy vào độ chặt chẽ của hệ thống chấm bài)
```

w4a8 so dao nguoc

```python
n=int(input('nhap n: '))

    
so_buoc = 0

while True:
        # Đảo ngược số bằng toán học
    tam_thoi = n
    so_dao = 0
    while tam_thoi > 0:
        chu_so_cuoi = tam_thoi % 10
        so_dao = (so_dao * 10) + chu_so_cuoi
        tam_thoi = tam_thoi // 10

        # Kiểm tra đối xứng
    if n == so_dao:
        break

        # Cộng dồn và tăng bước
    n = n + so_dao
    so_buoc += 1
print(so_buoc, n )
```

w4a13 close pair number

```python
n=int(input('nhap n: '))
m=int(input('nhap m: '))
#tong cac uoc cua mot so k
def tonguoc(k):
    tong=1
    for i in range(2, int(k**0.5)+1):
        if k%i==0:
            tong=tong+i
            if i*i !=k:
                tong=tong+k//i
    return tong

def kiemtrathanthiet(n,m):
    return tonguoc(n)==m and tonguoc(m)==n
print(kiemtrathanthiet(n,m))
```
