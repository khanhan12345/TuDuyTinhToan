w4a2 prime number

```python
n=int(input())
if n<=0:
    print('nhap lai')
if n==1:
     print('1 ko la so nguyen to')
if n>1:
    for i in range(2,(int(n**0.5))+1):
        if n%i==0:
            print(f'{n} ko la so nguyen to')
            break
    else:
            print(f'{n} la so nguyen to')
```
