bmi check

```python
def bmi_compare(w1, h1, w2, h2):
    if (w1/(h1*h1)) > (w2/(h2*h2)):
        return 1
    if (w2/(h2*h2)) > (w1/(h1*h1)):
        return 2
    if (w2/(h2*h2)) == (w1/(h1*h1)):
        return 0
    


assert bmi_compare(70, 1.75, 70, 1.75)==0
print('ok')
```

math grade

```python
def grade(s):
    if s>=9:
        return 'A+'
    elif s>=8.5:
        return 'A'
    elif s>=8:
        return 'B+'
    elif s>=7:
        return 'B'
    else:
        return 'E'
    
assert(grade(8.3))=='B+'
print('ok')
```

```python
def piecewise_mod3(n):
    if n%3==1:
        return 5*n+1
    if n%3==2:
        return 6*n+2
    if n%3==0:
        return 7*n+3
print(piecewise_mod3(n))
    
n=int(input())
```
