w5a1 ham max of two (a,b)

```python
a=int(input())
b=int(input())
def max_of_two(a,b):
    if max(a,b)==a:
        return a
    else:
        return b
print(max_of_two(a,b))
```

w5a2 viet ham hoan doi hai so nguyen ( def swap(a,b) -> tuple[int,int])

```python
a=int(input())
b=int(input())
def swap(a,b):
    temp=a
    a=b
    b=temp
    return(a,b)
print(swap(a,b))
```

w5a3 kiem tra so nguyen to

```python
n=int(input())
def check_prime_num(n):
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return 'khong la so nguyen to'
        else:
            return 'la so nguyen to'
print(f'{n} {check_prime_num(n)}')
```

w5a4 kiem tra mot so co la so hoan hao khong, kieu du lieu so nguyen
la so nguyen duong bang tong cac uoc so thuc su cua no(khong bao gom chinh no)

```python
n=int(input())

def check_perfect_num(n):
    temp=0
    for i in range(1,n):
        if n%i==0:
            temp=temp+i
    if temp==n:
            return 'la so hoan hao'
    else:
            return 'khong la so hoan hao'
print(check_perfect_num(n))
```

w5a5 viet mot ham dua vao mot list cac so nguyen va mot so nguyen duong k. tim va tra ve vi tri cua phan tu dau tien co gia tri k trong list so nguyen, neu khong co thi tra ve -1

```python
n=list(map(int, input().split()))
k=int(input())
def check_k_value(n,k):
    for i in range(len(n)):
        if n[i]==k:
            return f'vi tri thu {i+1}'
            break
    return '-1'
print(check_k_value(n,k))
```

w5a6 viet ham tinh giai thua

```python
n=int(input())
def factorial(n):
    temp=1
    for i in range(1,n+1):
        temp=temp*i
    return temp
print(factorial(n))
```

w5a7 gia lap may tinh bo tui

```python
num1, num2=map(int, input().split())
phep_toan=input()
def casio_emulator(num1, num2, phep_toan):
    if phep_toan=='-':
        kq=num1-num2
        return kq
    if phep_toan=='+':
        kq=num1+num2
        return kq
    if phep_toan=='*':
        kq=num1*num2
        return kq
    if phep_toan=='/':
        if num2==0:
            return 'khong thuc hien duoc phep chia cho 0'
        kq=num1/num2
        return kq
print(f'{casio_emulator(num1, num2, phep_toan):.2f}')
        
```

w5a9 viet ham tra ve tong cac chu so cua so nguyen duong n

```python
n=int(input())
def total(n):
    tong=0
    while n>0:
        chu_so=n%10
        tong=tong+chu_so
        n=n//10
    return tong
print(total(n))

```

w5a10 dang cau check

```python
#so sanh vi tri xuat hien dau tien
s=input()
t=input()
def dang_cau_check(s,t):
    if len(s)!=len(t):
        return False
    for i in range(len(s)):
        if s.find(s[i])!=t.find(t[i]):
            return False
    return True
print(dang_cau_check(s,t))
```

w5a12 divide_with_remainder(so chia, so bi chia)
output thuong, so du
(phep chia co du)

```python
dividend=int(input('nhap so bi chia: '))
divisor=int(input('nhap so chia: '))
def divide_w_remainder(dividend, divisor):
    if divisor ==0:
        return 'khong co phep chia cho 0'
    q=dividend//divisor
    r=dividend%divisor
    return (q,r)
print(divide_w_remainder(dividend, divisor))
```

w5a14 compute(a,b,op)

```python
a=float(input())
b=float(input())
op=input()
def compute(a,b,op):
    if op=='+':
        return a + b
    if op=='-':
        return a-b
    if op=='*':
        return a*b
    if op=='/':
        if b==0:
            return 'khong co phep chia cho 0'
        return a/b
print(compute(a,b,op))
```
