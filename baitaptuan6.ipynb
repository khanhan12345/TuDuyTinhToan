a6a8 nhap vao mot list cac chuoi, sau do dem so lan xuat hien cua tung chuoi va luu vao dictionary

```python
n=map(str, input().split())
def char_count(n):
    counts={}
    for ch in n:
        if ch in counts:
            counts[ch]=counts[ch]+1
        else:
            counts[ch]=1
    return counts
print(char_count(n))
```

w6a9 nhap vao hai dictionary. sau do gop lai thanh mot dictionary moi trong do neu mot key xuat hien o ca hai dict, gia tri se duoc cong lai, neu chi co mot dic thi giu nguyen. in ra dictionary moi theo thu tu tang dan cua key

```python
# BƯỚC 0: Đọc dữ liệu đầu vào
# Dữ liệu nhập vào có dạng {'a': 1, 'b': 2}. 
# Hàm eval() giúp máy tính hiểu đây là một Dictionary thực sự chứ không phải chuỗi văn bản thông thường.
chuoi_1 = input("Nhập dictionary 1: ")
chuoi_2 = input("Nhập dictionary 2: ")

# Xử lý trường hợp người dùng nhập trống, gán mặc định là {}
dict1 = eval(chuoi_1) if chuoi_1 else {}
dict2 = eval(chuoi_2) if chuoi_2 else {}

# BƯỚC 1: Chuẩn bị Dictionary kết quả
dict_ket_qua = {}

# Đưa toàn bộ các cặp key-value của dict1 vào dict_ket_qua
# Lệnh .items() giúp lấy ra từng cặp (key, value)
for key, value in dict1.items():
    dict_ket_qua[key] = value

# BƯỚC 2: Xử lý dict2
for key, value in dict2.items():
    # Kiểm tra xem key của dict2 đã tồn tại trong dict_ket_qua chưa
    if key in dict_ket_qua:
        # Nếu ĐÃ CÓ: Cộng dồn giá trị mới vào giá trị cũ
        dict_ket_qua[key] = dict_ket_qua[key] + value
    else:
        # Nếu CHƯA CÓ: Thêm cặp key-value mới này vào dict_ket_qua
        dict_ket_qua[key] = value

# BƯỚC 3: Sắp xếp theo thứ tự từ điển (A-Z) của Key
# Hàm sorted() sẽ lấy các keys và xếp theo thứ tự tăng dần
danh_sach_key_da_sap_xep = sorted(dict_ket_qua.keys())

# Tạo một Dictionary mới để lưu kết quả đã sắp xếp
dict_da_sap_xep = {}
for key in danh_sach_key_da_sap_xep:
    # Lấy value từ dict_ket_qua đưa sang dict_da_sap_xep theo đúng thứ tự key
    dict_da_sap_xep[key] = dict_ket_qua[key]

# BƯỚC 4: In ra kết quả
print(dict_da_sap_xep)
```

w6a10 Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên dương k. In ra 
list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là một tuple, hai số trong tuple 
có thể giống nhau, thứ tự xuất hiện trong danh sách của số thứ nhất nhỏ hơn thứ tự của 
số thứ hai. Các tuple trong list được sắp xếp theo thứ tự tăng dần của phần tử đầu tiên 
trong tuple. 

```python
# BƯỚC 1: Đọc dữ liệu đầu vào
lst1 = list(map(int, input().split()))
k = int(input())

# BƯỚC 2: Chuẩn bị một Set để lưu trữ kết quả
# Sử dụng Set để tự động loại bỏ các cặp tuple bị trùng lặp
ket_qua_set = set()

# BƯỚC 3: Vòng lặp tìm kiếm
# i chạy từ phần tử đầu tiên đến phần tử kế cuối
for i in range(len(lst1) - 1):
    
    # j chỉ chạy từ vị trí ngay sau i cho đến hết danh sách
    # Điều này đảm bảo: thứ tự xuất hiện của lst1[i] luôn đứng trước lst1[j]
    for j in range(i + 1, len(lst1)):
        
        # Nếu tổng bằng k
        if lst1[i] + lst1[j] == k:
            # Đóng gói 2 số thành 1 tuple
            cap_so = (lst1[i], lst1[j])
            
            # Thêm tuple vào set (nếu tuple đã tồn tại, set sẽ tự động bỏ qua)
            ket_qua_set.add(cap_so)

# BƯỚC 4: Sắp xếp và in kết quả
# Set không có thứ tự, nên ta phải ép kiểu về lại List để có thể sắp xếp
ket_qua_list = list(ket_qua_set)

# Hàm sort() khi áp dụng cho list chứa tuple sẽ tự động sắp xếp 
# tăng dần dựa trên phần tử đầu tiên của tuple.
ket_qua_list.sort()

print(ket_qua_list)
```

w6a12 tim phan tu xuat hien nhieu nhat

```python
# BƯỚC 1: Đọc dữ liệu đầu vào thành list các số nguyên
lst = list(map(int, input().split()))

# BƯỚC 2: Thống kê tần suất (Dùng Dictionary)
# Key là con số, Value là số lần xuất hiện của con số đó
dem_so = {}
for so in lst:
    if so in dem_so:
        dem_so[so] = dem_so[so] + 1
    else:
        dem_so[so] = 1

# BƯỚC 3: Tìm số xuất hiện nhiều nhất (Ưu tiên số nhỏ hơn nếu hòa)
so_nhieu_nhat = None        # Lưu trữ "Nhà vô địch"
max_lan_xuat_hien = 0       # Kỷ lục số lần xuất hiện hiện tại

# Duyệt qua từng cặp (số, số lần xuất hiện) trong cuốn sổ đếm
for so, so_lan in dem_so.items():
    
    # Trường hợp 1: Phá kỷ lục số lần xuất hiện
    if so_lan > max_lan_xuat_hien:
        max_lan_xuat_hien = so_lan
        so_nhieu_nhat = so
        
    # Trường hợp 2: Hòa kỷ lục. Ta phải xét thêm điều kiện "số nhỏ nhất"
    elif so_lan == max_lan_xuat_hien:
        if so < so_nhieu_nhat:
            so_nhieu_nhat = so

# BƯỚC 4: In ra kết quả
print(so_nhieu_nhat)
```

w6a13 dao nguoc key vao value cua dictionary

```python
# BƯỚC 1: Đọc dữ liệu đầu vào
# Lệnh split() tách chuỗi nhập vào thành một danh sách các từ.
# Ví dụ: Nhập "a 1 b 2" -> danh_sach_tu sẽ là ['a', '1', 'b', '2']
danh_sach_tu = input().split()

# BƯỚC 2: Chuẩn bị một dictionary rỗng để chứa kết quả sau khi đảo ngược
dict_dao_nguoc = {}

# BƯỚC 3: Duyệt qua danh sách để trích xuất từng cặp
# Dùng hàm range với bước nhảy (step) là 2.
# i sẽ lần lượt mang giá trị: 0, 2, 4...
for i in range(0, len(danh_sach_tu), 2):
    key_cu = danh_sach_tu[i]        # Phần tử đứng trước (Ví dụ: 'a')
    value_cu = danh_sach_tu[i + 1]  # Phần tử đứng ngay sau (Ví dụ: '1')
    
    # Thực hiện thao tác đảo ngược: 
    # Ép value_cu làm chìa khóa (Key), ép key_cu làm giá trị (Value)
    dict_dao_nguoc[value_cu] = key_cu

# BƯỚC 4: In ra dictionary mới đã được đảo ngược
print(dict_dao_nguoc)
```

w6a15 loc dictionary theo gia tri

```python
# BƯỚC 1: Đọc dữ liệu đầu vào
# Đọc dòng đầu tiên chứa các cặp key-value và tách thành danh sách
danh_sach_tu = input().split()

# Đọc dòng thứ hai chứa số nguyên k
k = int(input())

# BƯỚC 2: Chuẩn bị một dictionary rỗng để chứa kết quả sau khi lọc
dict_ket_qua = {}

# BƯỚC 3: Duyệt qua danh sách để trích xuất từng cặp
# Vẫn dùng hàm range với bước nhảy (step) là 2 giống bài trước
for i in range(0, len(danh_sach_tu), 2):
    key = danh_sach_tu[i]        # Phần tử đứng trước (Ví dụ: 'math')
    
    # Ép kiểu phần tử đứng sau thành số nguyên để có thể so sánh toán học
    value = int(danh_sach_tu[i + 1]) 
    
    # BƯỚC 4: Kiểm tra điều kiện lọc
    # Nếu value lớn hơn k thì mới thêm vào dictionary kết quả
    if value > k:
        dict_ket_qua[key] = value

# BƯỚC 5: In ra dictionary mới đã được lọc
print(dict_ket_qua)
```
